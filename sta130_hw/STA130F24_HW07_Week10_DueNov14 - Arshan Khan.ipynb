1. Simple linear regression only has 1 predictor variable which creates a relationship between only 2 variables which are the outcome and predictor. In multiple linear regression, there's a predictor variable with an indicator variable. This relates multiple variables (categories) with the outcome. The benefits that multiple linear regression has over simple linear regression includes being able to model and take into account more complex relationships as there can be many different factors to an outcome. This makes the model more accurate. Another benefit to multiple linear regression is the ability to model relationships between categorical and continuous variables. It is a lot more flexible than simple linear regression.

<br>

2. Using an indicator variable allows the option to compare categorical data with continuous data. This is done through indicator variables where each category has its own intercept parameter allowing for different outcomes based on which binary category it is that the data comes from. For example, if the indicator is 1, then the mean value of the category in question will be factored into the intercept for the model. A continuous variable cannot do this. A continuous variable does not change depending on the category and so categorical data cannot be factored into simple linear regression.

Below is the equation for simple linear regression.

$$ Y_i = \beta_0 + \beta_1 predictorA $$

However this is the form the equation takes using indicator variables

$$ Y_i = \beta_0 + 1(\beta_1) $$

3. Introducing a continuous variable to an indicator variable basically makes multiple linear equations with the linear regression model. Each category will have its own intercept and so the outcome will change depending on category. The data will be spread out between 2 different intercept values where both lines are parallel except that one is $ beta_0 $ distance away from the other vertically.

Below is the equation for simple linear regression.

$$ Y_i = \beta_0 + \beta_1 predictorA $$

However this is the form the equation takes using an indicator variable with a continuous variable

$$ Y_i = \beta_0 + \beta_1 predictorA + 1(\beta_2)  $$

If the indicator variable = 1 then the equation becomes:

$$ Y_i = \beta_0 + \beta_1 predictorA + \beta_2 $$

Where the slope becomes $ \beta_0 + \beta_2 $

4. Adding an interaction between a continuous and indicator variable results in having multiple intercepts and slopes as certain variables only affect the outcome based on the category. The effect this takes is that the predictor has an effect on the slope and vice versa if the indicator variable is equal to 1. This relates the categorical data to continuous data. This takes the form of the equation shown below:

$$ Y_i = \beta_0 + \beta_1 predictorA + 1(\beta_2) + \beta_3(predictorA * 1(\beta_2)) $$

5. The multiple linear regression model will get more complex with each indicator variable introduced. With multiple indicator variables, the model has multiple intercepts based on the categories and each category has a different line. The mean value of the outcome will change based on the category. This results in data whose mean is constantly changing according to the category showing the relationship between categories rather than the effect a continuous variable has on the outcome. This way we can very easily compare different categories with each other and create a trend between them. The form follows as shown.

$$ Y_i = \beta_0 + \beta_1 1(predictorA) + \beta_2 1(predictorB) $$

This results in a dummy variable which is the indicator. It is binary as either the indicator = 1 or 0. In other words, either the category is selected and the corresponding beta parameter is used in the outcome or it isn't.

Of course, an **interaction** is when the relationship of one **predictor variable** with the **outcome variable** depends on the value of another different **predictor variable**, so the impact of one **predictor variable** changes based on the presence or magnitude of another **predictor variable**. But are you sure you know what this means in the context of the **linear form** of a **multiple linear regression** model?_

> - _Imagine you're selling smoothies, the taste of your smoothie depends on the ingredients included in the smoothie, and there are two ingredients (bananas and strawberries) in the smoothie._
>     - _Adding more bananas into the smoothie will of course increase the "banana" flavor of the smoothie, and vice-versa for strawberries..._
>     - _But is this "banana" influence on the flavor always constant, or can it change depending on the absolute amount of strawberries in the smoothie?_ 
>     - _If the "banana" flavor influence is constant and does not depend on the  absolute amount of strawberries in the smoothie, then there is no **interaction** and the **linear form** of the model is $\beta_b b_i + \beta_s s_i$ and the model is said to be only **additive**._

> 
> - _So, if there is no **interaction**, then the effect of adding more bananas on the taste of the smoothie will always be the same, no matter how many strawberries you put in. > - _Or, on the other hand, if there is an **interaction**, then the effect of adding bananas (on the smootie flavor) will be different depending on how many strawberries there currently are in the smoothie: $\beta_b b_i + \beta_s s_i + \beta_{bs} (b_i\times s_i)$_

````python
import pandas as pd
import statsmodels.formula.api as smf

data = pd.read_csv('CSCS_data_anon.csv')

# Create binary variable for Sadness
data['Sadness'] = ((data['PSYCH_big_five_inventory_sad'] == 'Agree Strongly') | \
                  (data['PSYCH_big_five_inventory_sad'] == 'Agree a little')).astype(int)

# Remove missing values
data = data.dropna(subset=['DEMO_age'])
data = data.dropna(subset=['WELLNESS_self_rated_physical_health'])

# Fit logistic regression model
formula = 'Sadness ~ DEMO_age + C(WELLNESS_self_rated_physical_health)'
log_reg_model = smf.logit(formula, data=data).fit()
log_reg_model.summary()
````

$$ Y_i = \beta_0 + \beta_1 x_i + \beta_2 1(predictorA) + \beta_3 1(predictorB) + \beta_3 1(predictorC) + \beta_4 1(predictorD) $$

Thus for each category, we add the intercept of the category with $ \beta_0 $ and then add it to the age. Thus we can find out whether age and physical health have an effecton strongly feeling sadness. We can predict whether one feels strong sadness by finding out their self rated physical health and their age.

````python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

model7_linear_form_CS = 'HP ~ scale(center(Attack)) * scale(center(Speed))'
model7_linear_form_CS += ' * scale(center(Q("Sp. Def"))) * scale(center(Q("Sp. Atk")))'
# We DO NOT center and scale indicator variables
model7_linear_form_CS += ' + I(Q("Type 1")=="Normal")'
model7_linear_form_CS += ' + I(Q("Type 1")=="Water")'
model7_linear_form_CS += ' + I(Generation==2)'
model7_linear_form_CS += ' + I(Generation==5)'

model_formula = model7_linear_form_CS

# Number of iterations
n_iterations = 50

# Lists to collect in-sample and out-of-sample performance metrics
in_sample_rmse = []
out_sample_rmse = []
in_sample_r2 = []
out_sample_r2 = []

for i in range(n_iterations):
    # Randomly split data into train and test sets
    train_data, test_data = train_test_split(pokeaman_train, test_size=0.3)
    
    # Train the model on the training set
    model_spec = smf.ols(formula=model_formula, data=train_data)
    model_fit = model_spec.fit()
    
    # In-sample predictions (training set)
    train_pred = model_fit.predict(train_data)
    train_rmse = np.sqrt(mean_squared_error(train_data['HP'], train_pred))
    train_r2 = r2_score(train_data['HP'], train_pred)
    in_sample_rmse.append(train_rmse)
    in_sample_r2.append(train_r2)
    
    # Out-of-sample predictions (testing set)
    test_pred = model_fit.predict(test_data)
    test_rmse = np.sqrt(mean_squared_error(test_data['HP'], test_pred))
    test_r2 = r2_score(test_data['HP'], test_pred)
    out_sample_rmse.append(test_rmse)
    out_sample_r2.append(test_r2)

# Visualization
fig, axs = plt.subplots(2, 1, figsize=(12, 10))

# Plotting RMSE
axs[0].plot(in_sample_rmse, label='In-Sample RMSE', color='blue')
axs[0].plot(out_sample_rmse, label='Out-of-Sample RMSE', color='red')
axs[0].set_title('In-Sample vs. Out-of-Sample RMSE across Iterations')
axs[0].set_xlabel('Iteration')
axs[0].set_ylabel('RMSE')
axs[0].legend()

# Plotting R-squared
axs[1].plot(in_sample_r2, label='In-Sample R^2', color='blue')
axs[1].plot(out_sample_r2, label='Out-of-Sample R^2', color='red')
axs[1].set_title('In-Sample vs. Out-of-Sample R^2 across Iterations')
axs[1].set_xlabel('Iteration')
axs[1].set_ylabel('R^2')
axs[1].legend()

plt.tight_layout()
plt.show()
````
