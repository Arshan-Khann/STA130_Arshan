The difference between an attribute such as data.shape and a method such as data.describe() is attributes are simple characteristics you can run that hold a value or otherwise stores information. It doesn't require any input and doesn't do any computations.

In contrast, a method does do computations and it can be given an input, or arguments. Methods execute code in the module they are imported from. Attributes do not execute any code and just store information. That is the crucial difference.

```python
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-05-05/villagers.csv'
df = pd.read_csv(url)

# Basic summary of numerical columns
df.describe()
```

Above I have run df.describe() again as well as the full dataframe.

After some conversation with the ChatBot, it is clear that the 'count' statistic displays the total number of numerical values that are in row_n. 
The 'mean' statistic displays the mean or the average of the numerical values in row_n. It sums all values in row_n together then divides them by 'n' number of total terms, in this case being 391. 
The 'std' statistic calculates the standard deviation of the numeric values by square rooting the sum of the squared difference of each term and the mean, then dividing it by the total number of terms 'n'-1.
The 'min' statistic displays the minimum value found in the dataframe which in this case is 2.
The '25%' statistic displays the 25 percentile value or in other words, the value that marks the greatest of the bottom 25% of values after having all the values ordered from least to greatest.
The '50%' statistic displays the median, which is the middle value of the data when ordered from least to greatest. It shows the half-way point of the data.
The '75%' statistic shows the 75 percentile value, which is the value that appears at the 75% mark when the data is ordered from least to greatest. It tells us that this value is the greatest of the bottom 75% if values.
The 'max' statistic shows the greatest numerical value in the dataframe, in this case being 483.

df.describe() does not count non-numeric or missing values. That is why here it only shows data from row_n which is the only column that has traditional numeric values (birthday also has numeric values but those pertain to dates rather than numbers so they aren't considered 'numeric values').
