1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

Here's a summary of what we covered in this session:

1. **Classification Decision Trees**:  
   - They are used to classify data into distinct categories based on a set of input features. They split the data into branches to make a decision about the category.

2. **Metrics for Classification**:
   - **Accuracy**: Measures the proportion of correct predictions (both true positives and true negatives) among all predictions.
   - **Sensitivity**: Measures how well a test identifies actual positives (true positives), i.e., the proportion of sick people correctly identified in a medical test. It focuses on minimizing **false negatives** (Type II errors).
   - **Specificity**: Measures how well a test identifies actual negatives (true negatives), i.e., the proportion of healthy people correctly identified in a medical test. It focuses on minimizing **false positives** (Type I errors).
   - **Precision**: Measures how many of the predicted positive cases are truly positive (true positives). It focuses on minimizing **false positives** (Type I errors).

3. **Errors in Prediction**:
   - **False Positive (Type I Error)**: Occurs when a test incorrectly identifies a **healthy person as sick** or a **non-fraudulent transaction as fraudulent**.
   - **False Negative (Type II Error)**: Occurs when a test misses a **sick person** or a **fraudulent transaction**, incorrectly identifying them as healthy or non-fraudulent.

4. **Null Hypothesis for Metrics**:
   - For **Sensitivity**: The null hypothesis assumes that the test is no better than random guessing at identifying sick people (i.e., the person is healthy).
   - For **Precision**: The null hypothesis assumes that the test is no better than random at predicting true positives, meaning predictions of sick people or fraudulent transactions are no better than random guessing.

5. **Purpose of Sensitivity and Precision**:
   - **Sensitivity** is focused on **minimizing Type II errors** (false negatives), ensuring that sick people or fraud cases are not missed.
   - **Precision** is focused on **minimizing Type I errors** (false positives), ensuring that when a test predicts something as positive (e.g., sick, fraudulent), it is actually correct.

In essence, we discussed how classification metrics (Accuracy, Sensitivity, Specificity, and Precision) relate to errors in predictions (Type I and Type II errors), and how each metric serves different purposes in contexts like medical testing or fraud detection.

````python
# Prepare the target variable and feature variable
y = pd.get_dummies(ab_reduced_noNaN["Hard_or_Paper"])['H']  # Target: 1 if hardcover, 0 if paperback
X = ab_reduced_noNaN[['List Price']]  # Feature: List Price

# Split the data into training and testing sets (80/20 split)
ab_train, ab_test, y_train, y_test = train_test_split(X, y, train_size=0.8)

# Initialize the DecisionTreeClassifier with max_depth=2
clf = tree.DecisionTreeClassifier(max_depth=2)

# Train the model on the training data
clf.fit(ab_train, y_train)

# Make predictions on the test data
y_pred = clf.predict(ab_test)

# Evaluate the model: Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Display the confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

# Optional: Visualize the trained decision tree
dot_data = tree.export_graphviz(clf, out_file=None, 
                                feature_names=['List Price'],  
                                class_names=['Paperback', 'Hardcover'],
                                filled=True, rounded=True, special_characters=True)
graph = gv.Source(dot_data)

graph # View the tree using Graphviz
````

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
  
> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**. Fortunately, the "GridSearchCV" methodology mentioned in regards to finding the best `max_depth` setting for a tree is going to provide a general answer to the challenge of complexity and **overfitting** in **machine learning models** that is not too hard to understand (and which you might already have some guesses or a hunch about).

````python
import seaborn as sns

# Get feature importances and corresponding feature names
importances = clf2.feature_importances_
feature_names = clf2.feature_names_in_

indices = np.argsort(importances)[::-1]

importance_df = pd.DataFrame({
    "Feature": np.array(feature_names)[indices],
    "Importance": importances[indices]
})

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x="Importance", y="Feature", data=importance_df, palette="viridis")
plt.title("Feature Importances")
plt.show()
````
