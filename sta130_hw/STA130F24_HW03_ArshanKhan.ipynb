Below is the code provided by the ChatBot of the histograms.

````python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Load the penguin dataset
pingees = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Define a function to calculate necessary statistics and create histograms
def plot_flipper_length_histogram(penguins, species):
    data = penguins[penguins['species'] == species]['flipper_length_mm'].dropna()

    # Calculate statistics
    mean_flipper = data.mean()
    median_flipper = data.median()
    min_flipper = data.min()
    max_flipper = data.max()
    std_flipper = data.std()
    q1_flipper = data.quantile(0.25)
    q3_flipper = data.quantile(0.75)

    # Create the histogram using plotly.express
    fig = px.histogram(data, x=data, nbins=30, title=f'Flipper Length for {species}',
                       labels={'x': 'Flipper Length (mm)', 'y': 'Count'},
                       template='simple_white')

    # Customize the layout for simplicity and clarity
    fig.update_layout(
        title=dict(x=0.5, font=dict(size=18)),
        xaxis_title="Flipper Length (mm)",
        yaxis_title="Count",
        font=dict(size=14),
        bargap=0.05,
        margin=dict(t=60, l=50, r=50, b=50),
        showlegend=False
    )

    # Add vertical lines for mean and median with thicker lines and stronger contrast
    fig.add_vline(x=mean_flipper, line=dict(color='red', width=4), 
                  annotation_text="Mean", annotation_position="top right",
                  annotation=dict(bgcolor="white", bordercolor="red", borderwidth=2,
                                  font_size=12, font_color="red"))

    fig.add_vline(x=median_flipper, line=dict(color='purple', width=4), 
                  annotation_text="Median", annotation_position="top left",
                  annotation=dict(bgcolor="white", bordercolor="purple", borderwidth=2,
                                  font_size=12, font_color="purple"))

    # Add vertical rectangles for range (min-max), IQR, and mean +/- 2 std dev with lighter opacity
    fig.add_vrect(x0=min_flipper, x1=max_flipper, fillcolor="lightgray", opacity=0.15, 
                  line_width=0, annotation_text="Range", annotation_position="outside left",
                  annotation_font_size=10, annotation_font_color="gray")
    
    fig.add_vrect(x0=q1_flipper, x1=q3_flipper, fillcolor="lightblue", opacity=0.2, 
                  line_width=0, annotation_text="IQR", annotation_position="outside right",
                  annotation_font_size=10, annotation_font_color="blue")
    
    fig.add_vrect(x0=mean_flipper - 2*std_flipper, x1=mean_flipper + 2*std_flipper,
                  fillcolor="lightgreen", opacity=0.2, line_width=0, 
                  annotation_text="±2σ", annotation_position="outside right",
                  annotation_font_size=10, annotation_font_color="green")

    # Update x-axis range slightly beyond min and max for better visibility
    fig.update_xaxes(range=[min_flipper - 5, max_flipper + 5])

    # Show the plot
    fig.show(renderer='png')

# Plot histograms for each species
species_list = pingees['species'].dropna().unique()

for species in species_list:
    plot_flipper_length_histogram(pingees, species)
````

#### 2.
Kernel Density Estimation

Below is the code provided by the ChatBot.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the penguin dataset
penguins = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")

# Define a function to create KDE plots with statistical markings
def plot_kde_with_stats(penguins, species, ax):
    data = penguins[penguins['species'] == species]['flipper_length_mm'].dropna()

    # Calculate statistics
    mean_flipper = data.mean()
    median_flipper = data.median()
    min_flipper = data.min()
    max_flipper = data.max()
    std_flipper = data.std()
    q1_flipper = data.quantile(0.25)
    q3_flipper = data.quantile(0.75)
    
    # Plot the KDE using seaborn
    sns.kdeplot(data, fill=True, ax=ax, color='lightblue', label=f'{species} KDE')

    # Add vertical lines for mean and median
    ax.axvline(mean_flipper, color='red', linestyle='-', linewidth=2, label='Mean')
    ax.axvline(median_flipper, color='purple', linestyle='-', linewidth=2, label='Median')

    # Add shaded regions for range (min-max), IQR, and mean +/- 2 std dev
    ax.axvspan(min_flipper, max_flipper, color='lightgray', alpha=0.3, label='Range')
    ax.axvspan(q1_flipper, q3_flipper, color='lightblue', alpha=0.3, label='IQR')
    ax.axvspan(mean_flipper - 2*std_flipper, mean_flipper + 2*std_flipper, color='lightgreen', alpha=0.3, label='±2σ')
    
    # Label the plot
    ax.set_title(f'KDE for {species}', fontsize=14)
    ax.set_xlabel('Flipper Length (mm)', fontsize=12)
    ax.set_ylabel('Density', fontsize=12)
    
    # Add a legend to the current axis
    ax.legend(loc='upper left', fontsize=10)

# Create a row of three plots
species_list = penguins['species'].dropna().unique()
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Plot for each species
for ax, species in zip(axes, species_list):
    plot_kde_with_stats(penguins, species, ax)

# Adjust layout and display
plt.tight_layout()
plt.show()
````

#### 3.
Discussion about KDE vs. histograms.

My preference would vary between which type of visualization I prefer. If I only need a handful of visualizations then I vastly prefer KDE visualizations as they are more accurate to each datapoint and I find it much easier to read and interpret than the histograms. KDE visualizations are continuous which also makes more sense to find the mean and median of data as it can be more accurate than histograms which are discrete. It may be hard to pinpoint where exactly the median or mean is for the data since the data is sorted into bins which are arbitrary.

However, if I need to create a large amount of visualizations or if complexity needs to be taken into account, then I would prefer histograms as they are far less intensive on computers and easier to calculate.

````python
from scipy import stats
import plotly.express as px
import pandas as pd
import numpy as np
 
sample1 = stats.gamma(a=2,scale=2).rvs(size=1000)
fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data")
fig1.show(renderer="png")

sample1.mean()
np.quantile(sample1, [0.5]) # median

sample2 = -stats.gamma(a=2,scale=2).rvs(size=1000)
````

The mean and median are ways to measure data and what it may mean. The mean is the average of the data points which in this case would be closer to the peak of the graph. The median is the 50 percentile which means its the point at exactly 50% of the data when the data is ordered. Here, it would end up being right on the tail of the data here.

The mean and median would be in very different spots on this histogram due to the skewness. This is a right skewed histogram as the tail is on the right side and continues right. A left skewness would be on the left side. While the mean may not be affected by the skewness in this case as there is very few data points past the initial dropoff, the median is heavily affected by this data. The right skewness gives the median a very different value than the mean as the median isn't concerned with where there is more data or where the concentration of data is, it just shows the middle point of the data.

Skewness occurs due to the fact that a clear cutoff is not present. For example, we can see that the numbers are always positive with this data. This means there is a cutoff point at 0 where negative numbers are not considered. In this same way, there is not an upper bound for the data. Skewness occurs when there is a clear boundary on one side of the data but not the other. Here the boundary is for the left of the data while there isn't a clear bound for the right. We are considering things like outliers which which don't end up meaning much.

Provide a second version of the figure from the previous problem where you edit the fig = px.scatter() function from the Gapminder code so that x is "percent change", y is "rank", size is "percent", and color="sex", animation_frame is "year", and animation_group and hover_name are "name".
