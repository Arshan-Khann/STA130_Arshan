The linear regression model can be represented by the following equation $ Y_i = \beta_0 + \beta_1 x_i + \epsilon_i $
    <br>
    $ Y_i $ is the outcome based on the predictor variable which is represented by $ x_i $ which will be each data point.
    <br>
    $ \beta_0 $ is the intercept of the linear equation
    <br>
    $ \beta_1 $ is the slope of the linear equation
    <br>
    $ \epsilon_i $ is the error term or the amount of noise where $ \epsilon_i \sim \mathcal N(0, \sigma^2) $ which gives the data it's spread.
    
Combing these variables will give the above equation which represents a normal distribution since most of the outcomes  will centered at a point depending on the intercept and slope and predictor.

````python
import numpy as np
import matplotlib.pyplot as plt

# np.random.seed(130)

# Generate synthetic data
# Number of data points
n = 100

# Generate random x values between 0 and 10
x = np.linspace(0, 10, n)

# Define the true slope (beta_1), intercept (beta_0), and standard deviation of the noise (sigma)
beta_1 = 2.5  # Slope
beta_0 = 1.0  # Intercept
sigma = 2.0   # Standard deviation of the noise

# Generate normally distributed noise term epsilon_i ~ N(0, sigma^2)
epsilon = np.random.normal(0, sigma, n)

# Compute Y values based on the linear model: Y_i = beta_0 + beta_1 * x_i + epsilon_i
Y = beta_0 + beta_1 * x + epsilon

# Print the theoretical slope and intercept
print(f"Slope (beta_1): {beta_1}")
print(f"Intercept (beta_0): {beta_0}")

# Plot the data and the theoretical regression line
plt.scatter(x, Y, label='Data points')
plt.plot(x, beta_1 * x + beta_0, color='green', linestyle='--', label='Estimated line (without noise)')
plt.xlabel('x (Predictor)')
plt.ylabel('Y (Outcome)')
plt.title('Linear Regression Model (Without Fitting)')
plt.legend()
plt.show()
````

````python
import numpy as np
import pandas as pd
import plotly.express as px
import statsmodels.formula.api as smf  # This is to fit the model as a linear regression using smf

# Set seed for reproducibility
# np.random.seed(130)

# Generate synthetic data
# Number of data points
n = 100

# Generate predictor variable x values between 0 and 10
x = np.linspace(0, 10, n)

# Define the true slope (beta_1), intercept (beta_0), and standard deviation of the noise (sigma)
beta_1 = 2.5  # Slope
beta_0 = 1.0  # Intercept
sigma = 2.0   # Standard deviation to create noise

# Generate normally distributed noise term epsilon_i ~ N(0, sigma^2)
epsilon = np.random.normal(0, sigma, n)

# Compute Y values based on the linear model: Y_i = beta_0 + beta_1 * x_i + epsilon_i
Y = beta_0 + beta_1 * x + epsilon

# Create a pandas DataFrame to hold the simulated data
df = pd.DataFrame({'x': x, 'Y': Y})

# Fit a simple linear regression model using statsmodels
# We specify the formula 'Y ~ X', meaning we are modeling Y as a function of X
model = smf.ols('Y ~ x', data=df).fit()

# Print the summary of the fitted regression model
print(model.summary())  # Gives detals on the coefficients, parameters, p-value and the correlation (r-squared)
# model.summary().tables[1]  # Grabs the table at the specified index (second table of the summary)
# fitted_model.params  # Gives the estimated parameters for the regression model (intercept and slope)
# fitted_model.params.values  # Gives the estimated parameter values in a numpy array format
# fitted_model.rsquared  # Gives the r-squared value of the model (the proportion of variance in the outcome)

# Create the scatter plot with trendline using Plotly Express
# Adds the data to a legend and creates a scatter plot of the data with a trendline using ols
df['Data'] = 'Data' # hack to add data to legend 
fig = px.scatter(df, x='x',  y='Y', color='Data', 
                 trendline='ols', title='Y vs. x')

# This is essentially what above `trendline='ols'` does
fig.add_scatter(x=df['x'], y=model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

# Show the figure
fig.show(renderer='png')
````

````python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import plotly.express as px

# Set seed for reproducibility
# np.random.seed(130)

# Generate synthetic data
# Number of data points
n = 100

# Generate predictor variable x values between 0 and 10
x = np.linspace(0, 10, n)

# Define the true slope (beta_1), intercept (beta_0), and standard deviation of the noise (sigma)
beta_1 = 2.5  # Slope
beta_0 = 1.0  # Intercept
sigma = 2.0   # Standard deviation to create noise

# Generate normally distributed noise term epsilon_i ~ N(0, sigma^2)
epsilon = np.random.normal(0, sigma, n)

# Compute Y values based on the linear model: Y_i = beta_0 + beta_1 * x_i + epsilon_i
Y = beta_0 + beta_1 * x + epsilon

# Create a pandas DataFrame to hold the simulated data
df = pd.DataFrame({'x': x, 'Y': Y})

# Fit a simple linear regression model using statsmodels
# We specify the formula 'Y ~ X', meaning we are modeling Y as a function of X
model = smf.ols('Y ~ x', data=df).fit()

# Print the summary of the fitted regression model
print(model.summary())  # Gives detals on the coefficients, parameters, p-value and the correlation (r-squared)
# model.summary().tables[1]  # Grabs the table at the specified index (second table of the summary)
# fitted_model.params  # Gives the estimated parameters for the regression model (intercept and slope)
# fitted_model.params.values  # Gives the estimated parameter values in a numpy array format
# fitted_model.rsquared  # Gives the r-squared value of the model (the proportion of variance in the outcome)

# Create the scatter plot with trendline using Plotly Express
# Adds the data to a legend and creates a scatter plot of the data with a trendline using ols
df['Data'] = 'Data' # hack to add data to legend 
fig = px.scatter(df, x='x',  y='Y', color='Data', 
                 trendline='ols', title='Y vs. x')

# This is essentially what above `trendline='ols'` does
fig.add_scatter(x=df['x'], y=model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

# Find the range of x values for the predictor variable
x_range = np.array([df['x'].min(), df['x'].max()])
# This manually calculates a fixed linear regression based on the coefficients given 
# beta0 and beta1 are assumed to be defined
y_line = beta_0 + beta_1 * x_range
fig.add_scatter(x=x_range, y=y_line, mode='lines',
                name=str(beta_0)+' + '+str(beta_1)+' * x', 
                line=dict(dash='dot', color='orange'))

# Show the figure
fig.show(renderer='png')
````

The difference between the two lines of code is the method in which the trendline is created. 

The ols method will use a prebuilt functionality of plotly express to create a trendline based on the data. It simulates the random sampling variability along with the data as it calculates the coefficients in accordance to the data. It takes into account the sampling variability in the calculating. 

The manual calculation of the trendline (without the use of ols) will not take into account variability. It assumes the given coefficient values and creates an trendline to represent the data. As such, the trendline may not best represent the data as the given coefficient values estimate the actual coefficient values for the linear regression model.

By continually rerunning the code, we can visually see that the ols trendline is always changing due to our simulated random sampling as it follows the actual trend of the data. The dotted trendline (not using ols) stays the same since it is using the given coefficient values rather than calculating the coefficient values based on the data.

### 10. Let's now consider just the (*n=160*) long wait times (as specified in the code below), and write code to do the following:

````python
import plotly.express as px
import numpy as np

# Number of repetitions
n = 1000

# For reproducibility
# np.random.seed(130)

# So that we can append each bootstrapped and simulated slope to a list
bootstrapped_slopes = []
simulated_slopes = []

long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

# Number of repititions for bootstraping and simulated sampling
for i in range(n):
    # Bootstrapping 
    bootstrapped_sample = old_faithful[long_wait].sample(n=long_wait.sum(), replace=True)
    # Create a fitted model
    fitted_bootstrapped_model = smf.ols('duration ~ waiting', data=bootstrapped_sample).fit()
    # Append the slope parameter for the current sample
    bootstrapped_slopes.append(fitted_bootstrapped_model.params.values[1])

    # Simulation under the null hypothesis
    old_faithful_simulation = old_faithful[long_wait].copy()
    old_faithful_simulation['duration'] = 1.65 + 0*old_faithful_simulation.waiting + \
                                      stats.norm(loc=0, scale=0.37).rvs(size=long_wait.sum())
    # Create a fitted model
    fitted_simulated_model = smf.ols('duration ~ waiting', data=old_faithful_simulation).fit()
    # Append the slope parameter for the current sample
    simulated_slopes.append(fitted_simulated_model.params.values[1])

# Put the slope parameters in a np array for calculation of confidence interval and p_value
bootstrapped_slope_coefficients = np.array(bootstrapped_slopes)
simulated_slope_coefficients = np.array(simulated_slopes)

# Create sampling distribution of bootstrapped slope coefficients
fig1 = px.histogram(bootstrapped_slope_coefficients, nbins=30, opacity=0.75, 
                   labels={'value': 'Slope Coefficients', 'variable': 'Data'}, 
                   title="Bootstrapped Sampling Distribution of Slope Coefficients")
# Add vertical lines for the quantiles
fig1.add_vline(x=np.quantile(bootstrapped_slope_coefficients, [0.025, 0.5, 0.975])[0], line_dash="dash", line_color="blue", annotation_text="2.5% Quantile", annotation_position="top left")
fig1.add_vline(x=np.quantile(bootstrapped_slope_coefficients, [0.025, 0.5, 0.975])[1], line_dash="dash", line_color="red", annotation_text="Median", annotation_position="top left")
fig1.add_vline(x=np.quantile(bootstrapped_slope_coefficients, [0.025, 0.5, 0.975])[2], line_dash="dash", line_color="blue", annotation_text="97.5% Quantile", annotation_position="top left")

# Create sampling distribution for simulated slope coefficients
fig2 = px.histogram(simulated_slope_coefficients, nbins=30, opacity=0.75, 
                   labels={'value': 'Slope Coefficients', 'variable': 'Data'}, 
                   title="Simulated Sampling Distribution of Slope Coefficients", color_discrete_sequence=['orange'])

# print summary for slope and intercept
print(smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().summary().tables[1])

# Create a scatter plot with a linear regression trendline
fig3 = px.scatter(old_faithful[long_wait], x='waiting', y='duration', 
                 title="Old Faithful Geyser Eruptions for short wait times (>"+str(long_wait_limit)+")", 
                 trendline='ols')

# Calculate p_value
p_value = (np.abs(simulated_slope_coefficients) >= \
 smf.ols('duration ~ waiting', data=old_faithful[long_wait]).fit().params[1]).mean()

print("95% Confidence Interval: ", np.quantile(bootstrapped_slope_coefficients, [0.025, 0.975]))
print("P_value: ", p_value)

fig1.show(renderer='png')
fig2.show(renderer='png')
fig3.show(renderer='png') # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````

We can see from figure 1 which is the histogram of bootstrapped slope coefficients that 0 is not contained within the 95% confidence interval. I have also printed the exact quantile range as well and from there we can see that the 2.5% quantile gets very close to 0 however it does not reach zero.

The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) *hypothesis test*; but, we can do "two sample" hypothesis testing using a *permuation test*, or create a 95% *bootstrap confidence interval* for the difference in means of the two populations. 

### (A) Do a permuation test $\;H_0: \mu_{\text{short}}=\mu_{\text{long}} \; \text{ no difference in duration between short and long groups}$ by "shuffling" the labels
### (B) Create a 95% bootstrap confidence interval  by repeatedly bootstrapping within each group and applying *np.quantile(bootstrapped_mean_differences, [0.025, 0.975])* to the collection of differences between the sample means.    
### (a) Explain how the sampling approaches work for the two simulations.
### (b) Compare and contrast these two methods with the *indicator variable* based model approach used in Question 10, explaining how they're similar and different.<br>
    
<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _You'll need to create `for` loops for repeated (shuffling simulation) **permutation** and (subgroup) **bootstrapping**, where_
>
> - _"shuffling" for **permutation testing** is done like this `old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(n=len(old_faithful), replace=False).values)#.groupby('kind').size()`; then, the **mean difference statistic** is then calculated using `.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]` (so the **observed statistic** is `old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]`_
> 
>
> - _"two sample" **bootstrapping** is done like this `old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)#.groupby('kind').size()`; then, the **bootstrapped mean difference statistic** is then calculated using `.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]` (like the **observed statistic** except this is applied to the **bootstrapped** resampling of `old_faithful`)_

````python
# Number of repititions
n = 1000
perm_diffs = []  # List to append to

for i in range(n):
    shuffled_data = old_faithful.assign(kind_shuffled=old_faithful['kind'].\
                                      sample(n=len(old_faithful), replace=False).values)  #.groupby('kind').size()
    perm_diff = shuffled_data.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    perm_diffs.append(perm_diff)

# Compare observed statistic to the distribution of perm_diffs
observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
p_value = np.mean(np.abs(perm_diffs) >= np.abs(observed_diff))

# Create sampling distribution for simulated slope coefficients
fig2 = px.histogram(perm_diffs, nbins=30, opacity=0.75, 
                   labels={'value': 'Mean Differences', 'variable': 'Data'}, 
                   title="Shuffled Sampling Distribution of Mean Differences", color_discrete_sequence=['orange'])

fig2.show(renderer='png')
print("P-value:", p_value)
````

````python
# Number of repititions
n = 1000
bootstrapped_mean_differences = []  # List to append to

# np.random.seed(130)

for i in range(n):
    bootstrapped_data = old_faithful.groupby('kind').apply(lambda x: \
                                                           x.sample(n=len(x), replace=True)).reset_index(drop=True)
    bootstrapped_mean_difference = bootstrapped_data.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrapped_mean_differences.append(bootstrapped_mean_difference)

# Calculate the 95% confidence interval
confidence_interval = np.quantile(bootstrapped_mean_differences, [0.025, 0.975])

# Create sampling distribution of bootstrapped slope coefficients
fig = px.histogram(bootstrapped_mean_differences, nbins=30, opacity=0.75, 
                   labels={'value': 'Mean Differences', 'variable': 'Data'}, 
                   title="Bootstrapped Sampling Distribution of Mean Differences")
# Add vertical lines for the quantiles
fig.add_vline(x=confidence_interval[0], line_dash="dash", line_color="blue", annotation_text="2.5% Quantile", annotation_position="top left")
fig.add_vline(x=confidence_interval[1], line_dash="dash", line_color="blue", annotation_text="97.5% Quantile", annotation_position="top left")

fig.show(renderer='png')
print("Confidence Interval:", confidence_interval)
````

#### a)
We can see from the permutation test that the p-value we get is always 0. This is because of our assumption for the null hypothesis in that the short duration mean and long duration mean are equal. In other words, we are testing if the difference between the means is 0. As such, we get differences very close to 0 as per our assumption. As such, we cannot find any mean differences that are as or more extreme than the observed mean of about 2.3. and so our p-value is 0 rather than being a very small number. As such there is very strong evidence against the null hypothesis. This is in line with the indicator variable approach.

For the bootstrapping approach we do sampling with replacement to find means however the sample size is considered here unlike with the permutation test as the short and long duration means are all calculated at the original sample size. Thus there is no issue with this approach. We get a confidence interval that does not include a mean difference of 0.

#### b) 
We can see that with the indicator variable approach, the p-value was very small (it has been truncated, not 0) however the permutation test gave us a p-value of 0. While both these p-values are very close to each other, they are different. The methods do not have an identical result. It is also worth noting that the permutation test approach doesn't keep in mind the sample size of each category unlike the indicator variable approach. The long wait times have a greater sample size and so the sample will have a greater effect on the p-value.

We can also see that the confidence interval for our indicator variable approach is very close to our confidence interval of the bootstrapped difference in means. As we can see this method works very closly to the indicator variable approach however it is also not the exact same. The bootstrapping method has some variability to it as the confidence interval changes with each run of the code.
